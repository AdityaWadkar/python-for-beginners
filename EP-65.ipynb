```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Age'],bins=15)
plt.xlabel('age')
plt.ylabel('frequecy')
plt.title('Age data in graphical format')
plt.show()
```

```python
oc = df['Occupation'].value_counts()
```

```python
oc.plot(kind='bar')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title('Occupation data in graphical format')
plt.show()
```
