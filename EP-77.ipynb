```python
from bs4 import BeautifulSoup as bs
import requests as r
```

```python
url = "https://www.geeksforgeeks.org/implementing-web-scraping-python-beautiful-soup/"
```

```python
res = r.get(url)
```

```python
res.status_code
```

```python
soup = bs(res.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
print(soup.get_text())
```

```python
links = soup.find_all('a')
```

```python
for link in links:
    print(link.get('href'))
```

```python
tag_link = soup.select('.mobile-header-list-item')
for link in tag_link:
    print(link.prettify())
```
