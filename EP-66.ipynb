```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('sample_data.xlsx')
```

```python
df.sample(10)
```

```python
print(df.to_string())
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df,x='Occupation',y='Age',palette='Set2')
plt.xlabel("Occupation")
plt.ylabel("Age")
plt.title("boxplot demonstration")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(data=df,x='Occupation',y='Age',palette='Set3')
plt.xlabel("Occupation")
plt.ylabel("Age")
plt.title("violinplot demonstration")
plt.show()
```

```python
pivot = df.pivot_table(index='Occupation', columns='Age', aggfunc=len,fill_value=0)
plt.figure(figsize=(12,6))
sns.heatmap(data=pivot,cmap='Blues')
plt.xlabel("Age")
plt.ylabel("Occupation")
plt.title("heatmap demonstration")
plt.show()
```

Boxplot (boxplot): Shows the distribution of numerical data across several categories. Useful for identifying outliers and comparing distributions.

Violin Plot (violinplot): Combines a boxplot with a kernel density estimation. Provides a more detailed view of the distribution compared to a boxplot.

Heatmap (heatmap): Visualizes matrix-like data using colors. Useful for showing correlations between variables in a dataset.

Pairplot (pairplot): Plots pairwise relationships in a dataset. Useful for exploring relationships between multiple variables at once.

Jointplot (jointplot): Combines scatter plots and histograms to visualize the relationship between two variables along with their univariate distributions.

Barplot (barplot): Shows point estimates and confidence intervals as rectangular bars. Useful for comparing quantities across different categories.

Pointplot (pointplot): Shows point estimates and confidence intervals as individual points. Useful for visualizing trends over categorical variables.

Swarmplot (swarmplot): Displays categorical data with non-overlapping points. Useful for visualizing the distribution of data points along a categorical axis.

FacetGrid (FacetGrid): Allows for the creation of multi-plot grids based on the unique values of a variable. Useful for visualizing relationships across multiple subsets of the data.

Clustermap (clustermap): Hierarchically clusters rows and columns of a heatmap based on their similarity. Useful for identifying patterns and clusters in complex datasets.
