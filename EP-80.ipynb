```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np


name=[]
rating=[]
reviews=[]
service = []
employees = []
ownership = []
age = []
location = []


header = {'User-Agent': 'you user agent'}


for page in range(1,51):
    print(f"################# {page} #########################")
    webpage=requests.get(f"https://www.ambitionbox.com/list-of-companies?page={page}",headers=header).text
    soup=BeautifulSoup(webpage,'lxml')

    container = soup.find_all('div', class_='companyCardWrapper')
    for i in container:
        all_data = i.find('span',class_='companyCardWrapper__interLinking').text.strip()
        parts = all_data.split(" | ")

        if len(parts)==5:
            name.append(i.find('h2').text.strip())
            rating.append(i.find('span',class_='companyCardWrapper__companyRatingValue').text.strip())
            reviews.append(i.find('span',class_='companyCardWrapper__ActionCount').text.strip())

            service.append(parts[0])
            employees.append(parts[1])
            ownership.append(parts[2])
            age.append(parts[3].split()[0])
            location.append(parts[-1].split()[0])

        
df=pd.DataFrame({
   'name':name,
   'rating':rating,
   'reviews':reviews,
   'company_type':service,
   'Head_Quarters':location,
   'Company_Age':age,
   'No_of_Employee':employees,
   'Company_Sector':ownership
   })

    
```
