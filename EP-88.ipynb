# **Stop Words**

### Stop words are words that are removed from natural language data before or after processing because they are considered insignificant

 I am eating = I eating

 I was eating = I eating

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
spacy.lang.en.stop_words.STOP_WORDS
```

```python
text = "Natural language processing (NLP) is a subfield of artificial intelligence (AI) that allows computers to understand human language, both spoken and written. "
```

```python
doc = nlp(text)
```

```python
for t in doc:
  if not t.is_stop:
    print(t.text)
```
